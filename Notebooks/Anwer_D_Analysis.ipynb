```python
country_name =['Norway' , 'Denmark', 'Iceland', 'Switzerland', 'Finland', 'Netherlands', 'Canada', 'New Zealand', 'Sweden', 'Australia']
data =whr23_new_df.head(10)
```

```python
# Dataframe containing the top 10 countries for the plot
top_10_countries = ['Norway', 'Denmark', 'Iceland', 'Switzerland', 'Finland', 'Netherlands', 'Canada', 'New Zealand', 'Sweden', 'Australia']

# Plotting the data
data = whr23_new_df[['Economy', 'Family', 'Health', 'Freedom', 'Generosity', 'Corruption', 'Dystopia']].head(10)
ax = data.plot(
    kind='barh',
    stacked=True,
    figsize=(12.5, 7.5),
    xticks=np.arange(0, data.sum(axis=1).max() + 1)  # Adjust xticks to fit the range
)
ax.set_yticklabels(top_10_countries)

# Annotate with percentage labels
for index, row in data.iterrows():
    cumulative_value = 0
    for col in data.columns:
        cumulative_value += row[col]
        percentage = row[col] / row.sum() * 100
        if percentage > 0:  # Only add labels for non-zero values
            ax.text(
                cumulative_value - row[col] / 2,
                index,
                f'{percentage:.1f}%',
                va='center',
                ha='center',
                color='white' if percentage > 5 else 'black'  # Adjust label color for contrast
            )
# Setting the x-axis label
plt.xlabel("Happiness Score")

# Setting the plot title
plt.title('Happiness Score of the top 10 Countries')

# Adding the legend outside the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.savefig("Happiness_Top10Country.png")
plt.show()
```

#A grid of scatterplots showing the relationship between the 'Happiness Score' and 'Median Age' variables in the dataset whr23_new_df.

```python
sns.pairplot(data=whr23_new_df, height = 5, hue='Region',
                  x_vars=['Happiness Score'],
                  y_vars=['Median Age'])
plt.savefig("scatter_HapAge.png")
```

```python

slope, intercept = np.polyfit(whr23_new_df['Happiness Score'], whr23_new_df['Median Age'], 1)
# Create the pairplot
pair_plot = sns.pairplot(data=whr23_new_df, height=5, hue='Region', x_vars=['Happiness Score'], y_vars=['Median Age'])
# Add a general regression line and equation to all scatter plots
for ax in pair_plot.axes.flat:
    sns.regplot(data=whr23_new_df, x='Happiness Score', y='Median Age', scatter=False, color='red', ax=ax)
    ax.text(6, 18, f'y = {slope:.2f}x + {intercept:.2f}', color='red', fontsize=12)
plt.title('Happiness Score vs. Median Age with Line of Best Fit')
plt.savefig("Happiness_MedianAge.png")
plt.show()
```

#The pair plot showing the relationship between the 'Happiness Score' and 'Health' variables in the dataset whr23_new_df for all regions.

```python
sns.pairplot(data=whr23_new_df, height= 5, hue='Region',
                  x_vars=['Happiness Score'],
                  y_vars=['Health'])
plt.savefig("HapHealth.png")
```

```python
slope, intercept = np.polyfit(whr23_new_df['Happiness Score'], whr23_new_df['Health'], 1)

# Create the pairplot
pair_plot = sns.pairplot(data=whr23_new_df, height=5, hue='Region', x_vars=['Happiness Score'], y_vars=['Health'])
# Add a general regression line and equation to all scatter plots
for ax in pair_plot.axes.flat:
    sns.regplot(data=whr23_new_df, x='Happiness Score', y='Health', scatter=False, color='red', ax=ax)
    ax.text(6, 0.1, f'y = {slope:.2f}x + {intercept:.2f}', color='red', fontsize=12)
plt.title('Happiness Score vs. Health with Line of Best Fit')
plt.savefig("Happiness_Health.png")
plt.show()
```
